```python
def newtons(f, df, x0):
    EPS = 0.000001
    x = x0
    steps = 0
    while abs(f(x)) >= EPS:
        steps += 1
        x = x - f(x) / df(x)
    return x, f(x), steps

def f(x):
    return 2 * x**3 - 3 * x**2 - 2 * x - 4

def df(x):
    return 6 * x**2 - 6 * x - 2

root, val, steps = newtons(f, df, 3.0)
print root, val, f(root)
print 'total', steps, 'steps'
```

```python
def newtons(f, df, x0):
    EPS = 0.000001
    x = x0
    steps = 0
    while abs(f(x)) >= EPS:
        steps += 1
        x = x - f(x) / df(x)
    return x, f(x), steps

def f(x):
    return exp(2 * x) - x**2 - 1

def df(x):
    return 2 * exp(2 * x) - 2 * x

root, val, steps = newtons(f, df, 3.0)
print root, val, f(root)
print 'total', steps, 'steps'
```
